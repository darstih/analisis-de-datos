According to the article in the biomedical journal, the following is the 1998-2012 database:

```python
#Dataset revista biomédica
hom98_05 = '''24.715   100,0
60,3   25.855   100,0
63,5  28.789
100,0
69,8  30.103  100,0
72,2  31.807  100,0
75,1   25.614  100,0   59,4
22.369
100,0   51,3   18.875  100,0   42,9'''


hom06_12 = '''18.024
100,0
17.789   100,0
16.952   100,0
19.707   100,0
18439  100,0
16.685    100,0
15.747    100,0'''

hom1998_2012 = {}
hom98_05 = hom98_05.split() 
for i in range(0,len(hom98_05),3):
    a = hom98_05[i].split('.')
    hom1998_2012[int((i/3)+1998)] = a[0]+a[1]
hom06_12 = hom06_12.split()
for i in range(0,len(hom06_12),2):
    a = hom06_12[i].split('.')
    hom1998_2012[int((i/2)+2006)] = a[0]+a[1] if len(a[0])<=3 else a[0]

hom1998_2012 = pd.Series(hom1998_2012, name='Homicidios').astype(int)
plt.figure(figsize=(15,6))
plt.box(False)
plt.tick_params(bottom=False, left=False)
plt.xticks(range(1998, 2013), labels=hom1998_2012.index)
hom1998_2012.plot(color='#cd1600')
plt.xlabel('Año\nYear')
plt.ylabel('Cantidad de casos\nNumber of cases')
plt.title('Casos absolutos de mortalidad por homicidio en Colombia 1998-2012\nAbsolute cases of mortality from homicide in Colombia 1998-2012')
plt.grid(axis='y')
```
