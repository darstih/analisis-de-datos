## ¿Por qué es dificil ser cientifico de datos en colombia?<br>



El gobierno colombiano tiene diferentes entidades para brindar información a sus ciudadanos de forma libre, sin embargo, hay varias irregularidades con respecto a la calidad de los datos.<br>
    Podemos apreciar que en el año 2019 Colombia quedó en el puesto 96 en en indice de percepción de la corrupción de la ONG Transparency International<br>

<img src="https://lasillavacia.com/sites/default/files/indice.webp">

Este es el primer problema que nos aqueja la corrupción, más especificamente de los datos, <br><b>¿Podemos confiar en las bases de datos brindadas por las localidades y las entidades oficiales?</b><br> Según el indice de percepción anterior es que no muchos colombianos confían en estas entidades.<br><br> <b>¿Estas entidades nos esconden información?</b><br> Claramente una persona corrupta hará lo posible para esconder información de ojos curiosos o periodistas, sin embargo, no podemos especular.

Debemos brindar privacidad a todas las personas, entonces, supongo que es normal, que no se brinde información personal de alguna persona y esta bajo la Ley 1273 de 5 de enero de 2009<br>
